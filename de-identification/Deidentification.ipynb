```python
import pandas as pd

all_data = pd.read_csv('merged_data.csv')
all_data.head()
```

```python
all_data = all_data[['insurance_member_id', 'insurance_id', 'disease_treats', 'chemical_name', 'marketing_name', 'firstname', 'lastname', 'gender', 'address', 'zip', 'ethnicity', 'city', 'state']]
```

```python
all_data.head()
```

```python
all_data['insurance_member_id'] = all_data.apply(lambda x: '*' * len(x['insurance_member_id']), axis=1)
```

```python
all_data['insurance_id'] = all_data.apply(lambda x: '*' * len(str(x['insurance_id'])), axis=1)
```

```python
all_data['firstname'] = all_data.apply(lambda x: '****', axis=1)
all_data['lastname'] = all_data.apply(lambda x: '****', axis=1)
all_data['address'] = all_data.apply(lambda x: '****', axis=1)
all_data['city'] = all_data.apply(lambda x: '****', axis=1)
```

```python
all_data['zip'] = all_data.apply(lambda x: str(x['zip'])[:2] + '***', axis=1)
```

```python
def ethnicity(x):
    if 'white' in x['ethnicity'].lower() or 'asian' in x['ethnicity'].lower():
        return "{Asian, White}"
    if 'black' in x['ethnicity'].lower() or 'hispanic' in x['ethnicity'].lower():
        return "{Black, Hispanic}"
    return "{Other}"
all_data['ethnicity'] = all_data.apply(ethnicity, axis=1)
```

```python
def removeNaN(x):
    if not isinstance(x['disease_treats'], str):
        x['disease_treats'] = " "
    if not isinstance(x['chemical_name'], str):
        x['chemical_name'] = " "
    if not isinstance(x['marketing_name'], str):
        x['marketing_name'] = " "
    return x
all_data = all_data.apply(removeNaN, axis=1)
```

```python
all_data.to_csv('deindentified.csv', index=False)
```
