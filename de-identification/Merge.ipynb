```python
import pandas as pd


all_data = pd.read_csv("insurance_company.csv", delimiter=';')
```

```python
all_data.head()
```

```python
prescription = pd.read_csv("prescriptions.csv", delimiter=';')
prescription.head()
```

```python
insurance_health = pd.read_csv("insurance_health.csv", delimiter=';')
insurance_health.head(20)
```

```python
# prescription = prescription.groupby("insurance_id").apply(lambda x: " ".join(x['chemical_name']))
# prescription.head()
# data_all = data1.merge(data2, on = "pid", how = 'left')

merge1 = pd.merge(insurance_health, prescription, how='inner', on='perscription_id')
merge1 = merge1[['insurance_id', 'disease_id', 'disease_treats', 'chemical_name', 'marketing_name']]
def hh(x):
    x['chemical_name'] = " ".join(x['chemical_name'])
    x['marketing_name'] = " ".join(x['marketing_name'])
    return x
merge1 = merge1.groupby('insurance_id').apply(hh)
merge1 = merge1.drop_duplicates(['insurance_id'])
merge1 = merge1.sort_values('insurance_id')
merge1.head(20)
```

```python
merge1.head(20)
```

```python
merge2 = pd.merge(all_data, merge1, how='outer', on='insurance_id')
merge2 = merge2.sort_values('insurance_id')
merge2.head()
```

```python
merge2.fillna(' ')
merge2.to_csv('merged_data.csv', index=False)
```
